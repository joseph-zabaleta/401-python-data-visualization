### Visualization of the Dataset

```python
sns.relplot(x='horsepower', y='mpg', data=mpg)
```

```python
sns.lmplot(x='horsepower', y='mpg', data=mpg)
```

```python
sns.relplot(x='weight', y='mpg', data=mpg)
```

```python
sns.lmplot(x='weight', y='mpg', data=mpg)
```

### Conclusions:

Above is a simple but clear representation of some data from a built-in Dataset from Seaborn. This data set covers miles per gallon, for different model cars, from different countries, and of different sizes.

I choose to make two linear regression graphs to clearly points out these two things:  

1. First linear graph shows mpg vs horsepower. We can see that when horsepower increases the mpg's drops so there is a tradeoff fuel for power.  

2. I made a similiar finding with mpg vs weight. Weight increases , mpg decreses, and so when we increase horsepower we also increase the weight, which has a direct affect on the miles per gallon.
