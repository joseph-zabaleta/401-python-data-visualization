```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='ticks')
```

### Load data into Pandas DataFrame:

```python
exercise = sns.load_dataset('exercise')
```

### Visualization of the Dataset

```python
g = sns.FacetGrid(exercise, col='diet')
g.map(plt.hist, 'pulse')
```

```python
g = sns.FacetGrid(exercise, col='kind', hue='diet')
g.map(plt.scatter, 'time', 'pulse', alpha= 0.7)
g.add_legend()
```

```python
g = sns.FacetGrid(exercise, col='kind', height=4, aspect=0.5, hue='diet')
g.map(sns.barplot, 'time', 'pulse')
g.add_legend()
```

```python
g = sns.FacetGrid(exercise, hue='diet', palette='Set1', height=5, hue_kws={'marker': ['^', 'v']})
g.map(plt.scatter, 'time', 'pulse', s=100, linewidth=0.5, edgecolor='white')
g.add_legend()
```
